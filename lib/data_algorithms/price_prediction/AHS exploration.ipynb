```python
def howBuyColor(value):
    if value == "'-6'":
        return 'red'
    elif value == "' -9'":
        return 'blue'
    elif  value == "'1'":
        return 'yellow'
    elif value == "'2'":
        return 'orange'
    elif value == "'3'":
        return 'pink'
    elif value == "'4'":
        return 'green'
    elif value == "'5'":
        return 'purple'
    else:
        return 'cyan'
yearBuyData = data[data.YRBUILT > 2000]
yearBuyData.plot.scatter('YRBUILT', 'MARKETVAL', c=[howBuyColor(value) for value in yearBuyData['HOWBUY']])
```
